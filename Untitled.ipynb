```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, 'r')
```

```python
plt.subplot(1,2,1)
plt.plot(y,x,'b')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([.2,.5,.4,.3])
axes1.plot(x,y)
axes1.set_title('big')
axes2.plot(y,x)
axes2.set_title('small')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
plt.tight_layout()
axes[0].plot(x,y)
axes[0].set_title('1')
axes[1].plot(y,x)
axes[1].set_title('2')
```
